```python
import pandas as pd
import numpy as np
import warnings
import datetime
import statistics
warnings.simplefilter(action='ignore', category=FutureWarning)  
```

```python
df = pd.read_csv("milk_manual.csv")
df.head()
```

```python
df.pop("Unnamed: 1")
df = df.dropna()
df
```

```python
df[['Date', 'Time']] = df['created_at'].str.split('T', expand=True)
df[['Time', 'Standard']] = df['Time'].str.split('+', expand=True)

df.pop("Standard")
df
```

```python
df['Date'] = pd.to_datetime(df.Date)
df['Time'] = pd.to_datetime(df.Time)

df.dtypes
```

```python
df['Hour'] = df.Time.dt.hour
df['Mins'] = df.Time.dt.minute
df['Secs'] = df.Time.dt.second
df['Year'] = df.Date.dt.year
df['Month'] = df.Date.dt.month
df['Day'] = df.Date.dt.day
df['Time'] = df.Time.dt.time
df
```

```python
df.mean()
```

```python
 df.isnull().any().any()
```

```python
df['E1'] = 0
df['H1'] = 0
df['M1'] = 0
df['T1'] = 0
df['E2'] = 0
df['H2'] = 0
df['M2'] = 0
df['T2'] = 0


for i in range(3, len(df)-2):

    df.E1[i] = (( df.Ethanol[i-2] + df.Ethanol[i-1] + df.Ethanol[i] + df.Ethanol[i+1] + df.Ethanol[i+2])/5)
    df.H1[i] = (( df.Humidity[i-2] + df.Humidity[i-1] + df.Humidity[i] + df.Humidity[i+1] + df.Humidity[i+2])/5)
    df.M1[i] = (( df.Methane[i-2] + df.Methane[i-1] + df.Methane[i] + df.Methane[i+1] + df.Methane[i+2])/5)
    df.T1[i] = (( df.Temperature[i-2] + df.Temperature[i-1] + df.Temperature[i] + df.Temperature[i+1] + df.Temperature[i+2])/5)

for i in range(3, len(df)-2):

    df.E2[i] = (5 / ( (1/df.Ethanol[i-2]) + (1/df.Ethanol[i-1]) + (1/df.Ethanol[i]) + (1/df.Ethanol[i+1]) + (1/df.Ethanol[i+2])))
    df.H2[i] = (5 / ( (1/df.Humidity[i-2]) + (1/df.Humidity[i-1]) + (1/df.Humidity[i]) + (1/df.Humidity[i+1]) + (1/df.Humidity[i+2])))
    df.M2[i] = (5 / ( (1/df.Methane[i-2]) + (1/df.Methane[i-1]) + (1/df.Methane[i]) + (1/df.Methane[i+1]) + (1/df.Methane[i+2])))
    df.T2[i] = (5 / ( (1/df.Temperature[i-2]) + (1/df.Temperature[i-1]) + (1/df.Temperature[i]) + (1/df.Temperature[i+1]) + (1/df.Temperature[i+2])))
    
df.tail(20)    
```

```python
for i in range(1, len(df)):
    if (df.E1[i] == 0 or df.E2[i] == 0 ):
        df.E1[i] = df.Ethanol[i]
        df.E2[i] = df.Ethanol[i]
    if (df.H1[i] == 0 or df.H2[i] == 0 ):
        df.H1[i] = df.Humidity[i]
        df.H2[i] = df.Humidity[i]
    if (df.M1[i] == 0 or df.M2[i] == 0 ):
        df.M1[i] = df.Methane[i]
        df.M2[i] = df.Methane[i]
    if (df.T1[i] == 0 or df.T2[i] == 0 ):
        df.T1[i] = df.Temperature[i]
        df.T2[i] = df.Temperature[i] 

df
```

```python
df.pop("H1")
df.pop("T1")
df.pop("H2")
df.pop("T2")
```

```python
df['E3'] = 0
df['M3'] = 0
for i in range(3, len(df)-2):

    df.E3[i] = (( (df.Ethanol[i-2]-df.E1[i-2])**2 + (df.Ethanol[i-1]-df.E1[i-1])**2 + (df.Ethanol[i]-df.E1[i])**2 +(df.Ethanol[i+1]-df.E1[i+1])**2+(df.Ethanol[i+2]-df.E1[i+2])**2 )/5)
    df.M3[i] = (( (df.Methane[i-2]-df.M1[i-2])**2 + (df.Methane[i-1]-df.M1[i-1])**2 + (df.Methane[i]-df.M1[i])**2 +(df.Methane[i+1]-df.M1[i+1])**2+(df.Methane[i+2]-df.M1[i+2])**2 )/5)
```

```python
from sklearn import preprocessing

df['M4'] = 0
df['E4'] = 0
var1 = df.M1
a = preprocessing.normalize([var1])
var2 = df.E1
b = preprocessing.normalize([var2])

np.shape(b)
np.shape(a)
```

```python
a1 = a.T
b1 = b.T
np.shape(a1)
np.shape(b1)
```

```python
for i in range(1, len(df)):

        df.M4[i] = a1[i]
        df.E4[i] = b1[i]
df
```

```python
df['M5'] = 0

for i in range(1, len(df)):
        df.M5[i] = 1- ( ((df.M1[i]+df.M2[i])/2 - min(df.Methane)) / (500 - min(df.Methane)) ) - df.M4[i]
        
df
```

```python
df['E5'] = 0

for i in range(1, len(df)):
        df.E5[i] = 1 - ( ((df.E1[i]+df.E2[i])/2 - min(df.Ethanol)) / (500 - min(df.Ethanol)) ) - df.E4[i]
        
df     
```

```python
df['Quality'] = 0

for i in range(1, len(df)):
    df.Quality[i] = ((df.M5[i] + df.E5[i])/ 2)*10
    
df
```
